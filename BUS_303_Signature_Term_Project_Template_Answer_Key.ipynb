## Deliverable Three. 

### Comparing Starbucks and Dunkin' Donuts Menu for a Competitive Analysis.



**Task 3.1:**

- Create a horizontal bar chart comparing the Beverage Combination Portfolio Size of Starbucks and Dunkin' Donuts.    
- Add data labels showing the number of beverages on each bar.
- Include title, x-label, and y-label.
- Use the official exact colors of Starbucks (green) and Dunkin' Donuts (pink) for the chart.

One of the simplest ways companies compete is through product variety. This visualization compares the number of beverage combination offerings available at Starbucks and Dunkin' Donuts.

- Business Question: Which company provides customers with more beverage combination choices?

Starbucks

- What are some of the pros and cons of a variety of beverage combinations? 

A larger beverage portfolio may attract more customer segments but can also increase training, inventory, and operational complexity.

````python
# Task 3.1: Create a horizontal bar chart comparing the Beverage Combination Portfolio Size of Starbucks and Dunkin' Donuts.

import pandas as pd
import matplotlib.pyplot as plt

# Portfolio data
portfolio = pd.DataFrame({
    "Company": ["Starbucks", "Dunkin' Donuts"],
    "Beverage_combinations": [len(sbux4), len(dnkn4)]  # actual counts from your data
})

# Define the official colors for Starbucks and Dunkin' Donuts
starbucks_green = "#00704A"
dunkin_pink = "#FF5E9E"  # Dunkin’ bright pink
colors = [starbucks_green, dunkin_pink]

# Create a Horizontal Bar chart
ax = portfolio.plot(
    kind="bar",
    x="Company",
    y="Beverage_combinations",
    color=colors,
    figsize=(8,5),
    legend=False
)

# Add Titles & labels
plt.title("Beverage Combination Portfolio Size", fontsize=14, fontweight='bold')
plt.ylabel("Number of Beverage Combinations", fontsize=12)

# Add value labels on top of bars
for p in ax.patches:
    ax.annotate(str(p.get_height()), 
                (p.get_x() + p.get_width()/2., p.get_height()),
                ha='center', va='bottom', fontsize=12, fontweight='bold')

# Every Matplotlib chart has a frame around it called a “spine” (top, bottom, left, right). 
# The ax.spines setting hides unnecessary borders (top and right).
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.grid(axis='y', linestyle='--', alpha=0.5) # linestyle='--' adds dashed lines, which are lighter and less visually heavy than solid lines. 
# alpha=0.5 makes the lines semi-transparent, so they don’t overpower the chart.
# axis='y' adds grid lines that are drawn horizontally, along the y-axis, which helps read the values.

plt.show()
````

**Task 3.2:** 

- Create a bar chart comparing average beverage calories.

- Average Calories per Beverage
- Calculate the average calories for all beverages of each company.
- Create a bar chart to compare Starbucks vs Dunkin’ Donuts.
- Use brand colors consistently.
- Add value labels for the average calories on each bar.



````python
# Task 3.2: Create a bar chart comparing average beverage calories.

import pandas as pd
import matplotlib.pyplot as plt

avg_calories = pd.DataFrame({
    "Company": ["Starbucks", "Dunkin' Donuts"],
    "Average_Calories": [
        sbux4["Calories"].mean(),
        dnkn4["Calories"].mean()
    ]
})

starbucks_green = "#00704A"
dunkin_pink = "#FF5E9E"

ax = avg_calories.plot(
    kind="bar",
    x="Company",
    y="Average_Calories",
    color=[starbucks_green, dunkin_pink],
    figsize=(8,5),
    legend=False
)

plt.title("Average Beverage Calories", fontsize=14, fontweight="bold")
plt.ylabel("Calories", fontsize=12)

for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.1f}",
        (p.get_x() + p.get_width()/2., p.get_height()),
        ha="center",
        va="bottom",
        fontsize=11,
        fontweight="bold"
    )

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
````

````python
# Task 3.5: Compare healthier beverage choices.

sbux_healthy = len(
    sbux4[
        (sbux4["Calories"] < 100) &
        (sbux4["Sugar_g"] < 15)
    ]
)

dnkn_healthy = len(
    dnkn4[
        (dnkn4["Calories"] < 100) &
        (dnkn4["Sugars (g)"] < 15)
    ]
)

healthy = pd.DataFrame({
    "Company":["Starbucks","Dunkin' Donuts"],
    "Healthy_Beverages":[
        sbux_healthy,
        dnkn_healthy
    ]
})

ax = healthy.plot(
    kind="bar",
    x="Company",
    y="Healthy_Beverages",
    color=[starbucks_green,dunkin_pink],
    figsize=(8,5),
    legend=False
)

plt.title("Healthier Beverage Portfolio", fontsize=14, fontweight="bold")
plt.ylabel("Number of Beverages", fontsize=12)

for p in ax.patches:
    ax.annotate(
        str(int(p.get_height())),
        (p.get_x()+p.get_width()/2., p.get_height()),
        ha="center",
        va="bottom",
        fontsize=11,
        fontweight="bold"
    )

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
````
